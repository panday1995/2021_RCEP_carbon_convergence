```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns

%load_ext nb_black
```

```python
excel_path = r"D:\tencent_files\chrome_Download\Research\DEA\DEA_RCEP\Data"
excel = "Data.xlsx"
sheet = "per_capita_emission"

data = pd.read_excel(os.path.join(excel_path, excel), sheet_name=sheet)
data.head(2)
```

```python
sns.set_theme(style="ticks")

with sns.plotting_context(rc={"legend.fontsize": 12}):  # set legend font size
    sns.relplot(
        data=data,
        x="year",
        y="PC_emission",
        hue="Country",
        kind="line",
        palette=sns.color_palette("tab10", 15),  # set a colormap for line
        linewidth=3,  # set line width
        aspect=1.4,  # set the width/height ratio
        facet_kws=dict(sharex=False),
    )

    plt.xlabel("Year", fontsize=22)
    plt.ylabel("per capita emission (t)", fontsize=22)
    plt.tick_params(axis="both", which="major", labelsize=15)
    plt.xticks(range(2000, 2019, 2))
    plt.savefig(
        r"D:\tencent_files\chrome_Download\Research\DEA\DEA_RCEP\Graph\PC_emission.png",
        transparent=True,
        dpi=600,
        bbox_inches="tight",
    )
```
